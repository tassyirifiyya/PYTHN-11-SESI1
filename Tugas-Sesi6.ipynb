ubah 'Harley' pada NUM_BATH menjadi NaN lalu replace value NaN pada NUM_BATH menjadi median pada data number

```python
import pandas as pd
import numpy as np
```

```python
pr = pd.read_csv('property_data.csv')
pr
```

```python
cnt=0
for row in pr['NUM_BATH']:
    try:
        float(row)
    except ValueError:
        pr.loc[cnt, 'NUM_BATH']=np.nan
    cnt+=1
```

```python
pr
```

```python
pr['NUM_BATH'].fillna(pr['NUM_BATH'].median(),inplace=True)
pr
```

```python
#Kelompok 1
#Step1
import pandas as pd
import numpy as np

#Step2
pr = pd.read_csv('property_data.csv')
pr

#Step3 - mengubah string menjadi NaN
cnt=0
for row in pr['NUM_BATH']:
    try:
        float(row)
    except ValueError:
        pr.loc[cnt, 'NUM_BATH']=np.nan
    cnt+=1

#Step4
pr

#Step5 - Mengisi NaN dengan median
pr['NUM_BATH'].fillna(pr['NUM_BATH'].median(),inplace=True)
pr
```
