### Concat and Merge

```python
further_city_data = pd.DataFrame({'revenue':[7000,3400],
                                  'employee_count':[2,2]
                                 },index=['New York','Barcelona'])
further_city_data
```

```python
city_data
```

```python
all_city_data = pd.concat([city_data,further_city_data]) # Concat Vertical
all_city_data
```

```python
# Concat Horizontal
city_countries = pd.DataFrame({'country':['Holland','Japan','Holland','Canada','Spain'],
                              'capital':[1,1,0,0,0]
                              }, index=['Amsterdam','Tokyo','Rotterdam','Toronto','Barcelona'])
city_countries
```

```python
cities = pd.concat([all_city_data,city_countries],axis=1)
cities
```

```python
pd.concat([all_city_data,city_countries],axis=1,join='inner')
#join inner hanya menampilkan data yang ada
```

### Merge

```python
countries = pd.DataFrame({'population_mil':[17,127,37],
                         'continent':["Europe",'Asia','North America']},index=['Holland','Japan','Canada'])
countries
```

```python
countries2 = pd.DataFrame({'population_mil':[17,127,37],
                         'continent':["Europe",'Asia','North America'],
                          'country':['Holland','Japan','Canada']})
countries2
```

```python
cities
```

```python
#cities tabel 1, countries2 tabel 2
cities.merge(countries2, on ='country',how='inner') #default
```

```python
cities.merge(countries2, on ='country',how='left') #left
```

```python
cities.merge(countries2, on ='country',how='right') #right
```

## PR
Dengan menggunakan for loop dan try except
ubah 'Harley' pada NUM_BATH menjadi NaN lalu replace value NaN pada NUM_BATH menjadi median pada data number

```python
pr = pd.read_csv('property_data.csv')
pr
```

```python
cnt=0
for row in pr['NUM_BATH']:
    try:
        float(row)
    except ValueError:
        pr.loc[cnt, 'NUM_BATH']=np.nan
    cnt+=1
```

```python
pr.NUM_BATH.info()
print(type('NUM_BATH'))
```

```python
pr
```

```python
pr['NUM_BATH'].fillna(pr['NUM_BATH'].median(),inplace=True)
pr
```
