```python
x = [8, 1, 2.5, 4, 28]
#manual
def median_function(x):
    n = len(x)
    if n%2 == 1: #pengecekan apakah jlh data ganjil atau genap
        median_ = sorted(x)[round(0.5*(n-1))]
    else:#Genap
        x_ord, index = sorted(x), round(0.5*n)
        median_=0.5 * (x_ord[index-1]+x_ord[index])
    return median_
```
