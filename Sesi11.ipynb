## Simple Linear Regression

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1,1))
y = np.array([5, 20, 14, 32, 22, 38])

print(x)
print(y)
```

```python
np.array([5, 15, 25, 35, 45, 55]).reshape((-1,1))
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
#Makin dekat 1 makin bagus
r_sq = model.score(x,y)
r_sq
```

```python
print('intercept :', model.intercept_)
print('slope :', model.coef_)
```

# Advanced ... StatsModels

```python
import statsmodels.api as sm
```

```python
x = np.array([[0,1],
             [5,1],
             [15,2],
             [25,5],
             [35,11],
             [45,15],
             [55,34],
             [60,35]])

y = np.array([4, 5, 20, 14, 32, 22, 38, 43])

print(x,'\n', y)
```

```python
x = sm.add_constant(x)
x
```

```python
model_stat = sm.OLS(y,x)
```

```python
results = model_stat.fit()
```

```python
print(results.summary())
```

```python
print('rsquared:', results.rsquared)
```

```python
results.params
```

```python
res_summary = results.summary()
res_summary
```

```python
res_summary.tables[1]
```
